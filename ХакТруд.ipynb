```python
dataset = pd.read_csv('/content/drive/MyDrive/vvss.csv', header=0)
dataset.head()
```

```python
#Удаляем нулевые столбцы
dataset = dataset.drop('age', 1)
dataset = dataset.drop('elementaryEducation', 1)
```

```python
#Сколько резюме без вышего образования и доп.образования
((dataset['primaryEducation'] == '{}') & (dataset['additionalEducation'] == '{}')).sum()
```

```python
#Удаляем резюме без вышего образования и доп.образования
dataset = dataset.drop(dataset[(dataset['primaryEducation'] == '{}') & (dataset['additionalEducation'] == '{}') ].index)
```

```python
#Сколько резюме с нужным городом
(dataset['areaId'] == 2).sum()
```

```python
#Удаляем резюме с другими городами
dataset = dataset.drop(dataset[(dataset['areaId'] != 2)].index)
```

```python
#Сколько резюме с опытом работы: 1 и более лет
((dataset['workExperienceMonthCount'] >= 12)).sum()
```

```python
#Удаляем резюме с опытом работы: менее 1 года
dataset = dataset.drop(dataset[dataset['workExperienceMonthCount'] < 12 ].index)
```

```python
#Сколько резюме с направлением подготовки: техническое
searchfor = ['технич', 'Технич']
dataset['primaryEducation'].str.contains('|'.join(searchfor)).sum()
```

```python
#Копируем данные резюме в отдельный датасет
searchfor = ['технич', 'Технич']
datasetTech = dataset[dataset['primaryEducation'].str.contains('|'.join(searchfor))]
```

```python
#Сколько резюме с нужными скиллами
searchfor = ['Windows', 'AD', 'Exchange', 'SQL', 'DHCP', 'DNS', 'VMWare', 'Cisco', 'Zabbix']
datasetTech['experiences'].str.contains('|'.join(searchfor)).sum()
```

```python
#Копируем данные резюме с нужными скиллами
datasetTechSkills = datasetTech[datasetTech['experiences'].str.contains('|'.join(searchfor))]
```
